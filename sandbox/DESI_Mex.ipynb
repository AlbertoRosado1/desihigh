
Now as part of [Dark Energy Spectrocopic Instrument](https://www.desi.lbl.gov/), I am involved mostly in Clustering and Cosmological Simulations working groups, mainly working in the  "Mock challenge" team. Our goal is to prepare the mock catalogs and to design the validation test for DESI Analysis pipeline(set of codes for analising the samples of galaxies and quasars). This "pipeline" will be used by DESI team for analizing the two consolidated observables for stuying [dark energy](https://www.desi.lbl.gov/cosmology-and-dark-energy/): the [Baryon Acoustic Oscillations](https://en.wikipedia.org/wiki/Baryon_acoustic_oscillations) and the [Redshift Space Distortions](https://en.wikipedia.org/wiki/Redshift-space_distortions).
