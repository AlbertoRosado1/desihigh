```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from   astropy.table import Table
from   tools.legacy import cutout
```

# Having a gas

```python
dat = pd.read_csv('prep/tsz/andes_tsz.txt', sep=',')
```

```python
dat = Table(np.c_[dat.iloc[:,4], dat.iloc[:,5]], names=['RA', 'DEC'])
```

```python
dat
```

```python
fig, axes = plt.subplots(1, len(dat), figsize=(50,10))

for i, ax in enumerate(axes):
  cutout(ax, dat['RA'][i], dat['DEC'][i])
```
