```python
favoriteFruits = numpy.array(["mango", "lychee", "grapefruit", "papaya", "mangosteen", "guava"])
print(favoriteFruits)
```

It doesn't seem like much at first, but here's where we get to use our new powers! NumPy also has a feature that allows you to go through an array while skipping certain increments. For example, if I want to go back to my list of favorite fruits and only print out every other fruit, (the output should be ["mango", "grapefruit", "mangosteen"]), I could do that easily using the power of `numpy.arange`! (And yes, "arange" is indeed spelled with only one "r".)

```python
# favoriteFruits = numpy.array(["mango", "lychee", "grapefruit", "papaya", "mangosteen", "guava"])

# We'll get into how this all works very shortly!
for i in numpy.arange(0, 5, 2):
    print(favoriteFruits[i])
```

"numpy.arange" creates an array of numbers by taking in 3 values: the starting value, ending value, and the "step" or amount to increase by each time. Written out, it would look something like this: `numpy.arange(start, end, step)`. In the cell below, we can see exactly what numpy.arange does in action. Try playing around with the 3 numbers a bit. *Can you get it to print out [2, 5, 8, 11]?* (Note: the "start" number is included in the array, but it will stop right before the "end" number, which is why 7 doesn't get printed out too.)

```python
print(numpy.arange(1, 7, 2))
# What can you change the numbers 1, 7, 2 into so that it will print out [2, 5, 8, 11] instead?
```

Now it's time to apply this to other arrays! We can use a "for" loop to access the elements in an array using their indices, which will come from the sequence of numbers generated by numpy.arange. In the earlier example, `numpy.arrange(0, len(favoriteFruits), 2)` resulted in the array [0, 2, 4], which are the indices of "mango", "grapefruit", and "mangosteen" respectively, so they are the ones that got printed out.
