## II. A BRIEF HISTORY OF SPECTROSCOPY

*Sir Isaac Newton, who allowed sunlight to pass through a round hole in a screen, interposed a prism in the path of the beam of light, and projected a band of colors on the wall, is sometimes referred to as the father of spectroscopy. In 1664, Newton was able to spread this band of light by interposing lenses of various kinds.  However, the real contribution of Newton in the above experiment was the demonstration of the composite nature of light.*

*Herschel in the l800s, from his studies of the spectral distribution of heat that the maximum lies beyond the red end, the infrared spectrum. But it was not until 1802, when Thomas Young discovered his wave theory of light, that any
positions in the spectrum were known.  Young was able to calculate, with reasonable accuracy, the seven colors of Newton.* 

```python
Image('desihigh/images/Newton.jpeg', width=250)
```

*In 1814, Joseph von Fraunhofer recognized spectrum lines in the sun.  He identified 324 dark lines. which are the well-known "Fraunhofer Lines". The instrument was exceedingly simple; he allowed sunlight to pass through a slit and used a telescope instead of a screen.  Wollaston, twelve years prior to Fraunhofer, saw these dark lines, but thought they separated the colors, or the continuum between the Fraunhofer lines.*

```python
Image('desihigh/images/fraunhofer.jpeg', width=400)
```

*Fraunhofer and his associates found that certain bright lines in flame spectra coincided with these dark lines in the sun.  Fraunhofer was a great contributor to spectroscopy, for along with the above he measured the Sodium D line, and invented the grating, which meant that he understood diffraction. However, the coincidence of the lines in the flame spectrum with the Fraunhofer lines was not explained until 1859. by G. R. Kirchhoff. He proved by theory and experiment, using D1 DZ lines of sodium, that each atom is unique. Spectroscopy was introduced to astrophysics and chemical analysis in 1861 by Kirchhoff and Bunsen. They compared flame spectra of very pure elements with the solar spectra.  This led to the discovery of rubidium and cesium.*  

```python
Image('desihigh/images/rb.jpg', width=200)
```

*These two elements were present in the sun's spectrum so evidently that Kirchhoff and Bunsen felt they could reproduce these lines in the laboratory. Hydrogen was found in the atmosphere of the sun in 1862 by A. J. Angstrom, who published the first solar standards in 1868. His measurements of one thousand Fraunhofer lines were made in units of $10^{-8}$ cm. This unit is used today and is known as the Angstrom.*  

*Several cadmium lines were measured in 1892 by A. A. Michelson.  The fact that Michelson was measuring the spectrum of the element that was to contain the primary standard is _not_ his great contribution.  Michelson invented the interferometer, which was to set off a new era and provide even greater accuracy for the spectroscopist. In 1905, Benoit Fabry and Perot, with an interferometer made by Fabry and Perot, measured the red line of cadmium. They obtained, the wavelength, $6,438.4696$ Angstroms, using Benoit's standard meter.  This line has been accepted as the primary standard.*

*Thousands of wavelengths have been found relative to this standard, and many of these have been internationally accepted as standards.  The Eleventh General Conference on Weights and Measures, which was held in Paris, France, December 1960, replaced the platinum-iridium meter bar with the orange-red line of krypton 86. source is operated at the triple point temperature of nitrogen, 63$\deg$K. The wavelength is 6057.8021\AA and the meter is defined as
1,650,763.73 waves per meter.*

*About the beginning of the 20th century, the spectroscopist realized that wavelengths of the lines of any spectrum whatsoever, could be calculated by interpolation between secondary standards coincident on the same spectrogram.  Thus, the spectroscopist of the old era could foresee that the number of determined wavelengths would grow and intensify the knowlege of chemical structure, atomic and nuclear structure and extend through identification of the elements the* **observational information about our universe.**

In total, he recorded a total of 5,900 Titanium lines over a total of 118 pages.  Noone said a P.h.D would be easy! 
By the end, he would have been able to summaris his hard work in his highly precise Titanium spectrum: 

```python
# Fetch a Titanium spectrogram.  
dat = pd.read_csv(resource_filename('desihigh', 'dat/titanium.dat'), sep='\s+', comment='#', names=['Intensity', 'Wavelength'], usecols=[0,1])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,5))

for wave in dat['Wavelength']:
  # color = [r, g, b]
  color = wavelength_to_rgb(wave / 10.)       
  ax.axvline(x=wave, c=color, lw=0.4)

ax.set_xlabel('Vacuum wavelength [nanometers]')
ax.set_title('Spectrogram of Titanium')
```

Why? Dr Banks explains:

*The spectrum of titanium is characteristic of elements with a many electron system.  There is no sharp demarcation ... series and the spectrum is complex. On the other hand, the titanium spectrum lends itself well to analysis and prediction.  Thus the purpose of this investigation is to verify predicted levels, to determine new levels from lines hitherto not seen or recorded and to improve upon the past analysis of the neutral spectrum of titanium
by increasing the overall accuracy.*

*The titanium atom has an argon core of eighteen electrons and an atomic number of twenty two. There are four electrons in the outermost shell, two in the 3d and two in the 4s shell.  The ionization potential, energy for the removal of one electron from the outer shell, is 6.83 ev.*

*The astrophysical importance of this investigation is three-fold. First, titanium has been found to be present in* **sunspots** *and in the* **cooler stars**. *In the long period variables, such as O Cett, it has been noted that as the intensity of the light fades, the titanium lines become stronger. Secondly, in the cooler stars of class K2, KS, MO, MS, and M8 a large number of titanium lines has been observed.  Thirdly, the titanium spectrum analysis will facilitate the identification of Fraunhofer lines and other faint lines in the solar spectrum.* **After iron, titanium ranks second in the abundance of lines in the solar spectrum.** *

DESI Instrument scientist, Dr Julien Guy, employs these same standards to extract a precise spectrum from DESI, employing Mercury, Argon, Neon and Cadmium, in stead of Titanium.  Resulting in a calibration of the wavelength to 0.18 Angstroms, in contrast Dr Banks made highly precise 0.023 Angstrom measurements.
